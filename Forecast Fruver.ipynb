# Forecast Kampos 175

```python
forecast1 = """SELECT CITY, WAREHOUSE, RETAIL_ID, PRODUCT_NAME, FORECAST_1 as Lunes, FORECAST_2 as Martes, FORECAST_3 as Miercoles, FORECAST_4 as Jueves, FORECAST_5 as Viernes, FORECAST_6 as Sabado, FORECAST_7 as Domingo, FORECAST_8 as Lunes_1, FORECAST_9 as Martes_1, FORECAST_10 as Miercoles_1, FORECAST_11 as Jueves_1
from fivetran.turbo_turbo_supply.roq_direct_supply_sync  
where country = 'CO' 
AND date = CURRENT_DATE() - INTERVAL '1 DAY'
  AND IS_PO_DAY = 'true'
AND provider_id IN (175)"""
```

```python
df_forecast1 = pd.read_sql(forecast1, conn)
df_forecast1.head(2)
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Simulación de lectura de datos desde SQL (reemplaza con tu propio código)
# df_forecast1 = pd.read_sql(forecast1, conn)

# Crear una fecha actual formateada
fecha_actual = datetime.now().date()

# Crear un DataFrame con las fechas en las filas, comenzando desde fecha_actual + 1
num_columnas = 7
columnas_fechas = [(fecha_actual + timedelta(days=i + 1)).strftime('%d-%m-%Y') for i in range(num_columnas)]
df_fechas = pd.DataFrame({'Fechas': columnas_fechas})

# Crear un archivo de Excel con el nombre personalizado
nombre_archivo = f'Forecast-Kampos-{fecha_actual.strftime("%d-%m-%Y")}.xlsx'

# Crear un escritor de Excel
with pd.ExcelWriter(nombre_archivo, engine='xlsxwriter') as writer:
    # Transponer el DataFrame de fechas para que estén en filas
    df_fechas = df_fechas.transpose()
    
    # Escribir los datos del DataFrame de fechas en la hoja 'Sheet1' desde la fila 1, empezando desde la columna F (índice 5)
    df_fechas.to_excel(writer, sheet_name='Sheet1', startrow=0, startcol=4, header=False)
    
    # Escribir los datos del DataFrame en la hoja 'Sheet1' desde la fila 2
    df_forecast1.to_excel(writer, sheet_name='Sheet1', startrow=1, index=False)

print(f'Archivo de Excel "{nombre_archivo}" creado con éxito.')
```

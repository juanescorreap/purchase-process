```python
import pandas as pd

# Leer el archivo
df = pd.read_csv(r"C:\Users\esteban.correa\Downloads\Libro1.csv")

# Crear listas únicas de WH y SKU
warehouse_ids = df['WH'].unique()
product_ids = df['SKU'].unique()

# Crear DataFrames a partir de las listas únicas
warehouse_df = pd.DataFrame(warehouse_ids, columns=['warehouse_id'])
product_df = pd.DataFrame(product_ids, columns=['product_id'])

# Agregar la columna 'key' para el cross join
warehouse_df['key'] = 1
product_df['key'] = 1

# Realizar el cross join y eliminar la columna 'key'
combined_df = pd.merge(warehouse_df, product_df, on='key').drop('key', axis=1)

# Guardar el DataFrame resultante en un archivo CSV
csv_filename = r"C:\Users\esteban.correa\Downloads\warehouse_products.csv"
combined_df.to_csv(csv_filename, index=False)
```
